Code below is iterating over datan object and provides 
the Java Variable expressions for every column name, based on the columns's data type 

```python
tmap=[]
for row in datan.itertuples():
    if row.DATA_TYPE=='date' or row.DATA_TYPE=='datetime' or row.DATA_TYPE=='timestamp': 
        tmap.append('(rs.{0}!=null&&row1.{0}!=null)?(TalendDate.compareDate(rs.{0},row1.{0},"yyyy-MM-dd HH")==0?1:0):1'.format(row.COLUMN_NAME))
    else:
        tmap.append('(rs.{0}!=null&&row1.{0}!=null)?(rs.{0}.compareTo(row1.{0})==0?1:0):1 '.format(row.COLUMN_NAME))
datan["tmap"]=tmap
print(datan)
```
